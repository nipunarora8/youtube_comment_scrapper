# Finding Youtube Comments using Youtube API

<p>This is a simple program to find comments of a particular video on youtube using YouTube API. <br>
You just need to provide the link and job is done.<br>
The output of this program is a csv file that comprises of Video Name, Channel Name, Name of the person writing the comment, Comment Id, Comment, Date, Likes</p>

```python
#Youtube API key

api_key ="*************************************"
```

<p>You can get your own google api key by just signing up in www.console.developers.google.com and choosing youtube data from the libraries menu, create a new project and generate keys accordingly</p>

```python
from apiclient.discovery import build
```

```python
yt = build('youtube','v3',developerKey=api_key)
```

```python
#Enter Video Link here

query=input("Enter Video Link: ")
```

```python
#Finding video id

req=yt.search().list(q=query,part='snippet',type='video',maxResults = 1).execute()
```

```python
videoid=req['items'][0]['id']['videoId']
```

```python
#Finding comments for that video id

comments = yt.commentThreads().list(
                    part="snippet",
                    videoId = videoid,
                    maxResults = 200, # Only take top 100 comments...
                    order = 'relevance', #... ranked on relevance
                    textFormat = 'plainText',
                    ).execute()
```

```python
comment = []
comment_id = []
comment_name=[]
comment_date=[]
comment_likes=[]
for i in comments['items']:
    comment_id.append(i['snippet']['topLevelComment']['id'])
    comment_name.append(i['snippet']['topLevelComment']['snippet']['authorDisplayName'])
    comment.append(i['snippet']['topLevelComment']['snippet']['textDisplay'])
    comment_date.append(i['snippet']['topLevelComment']['snippet']['updatedAt'])
    comment_likes.append(i['snippet']['topLevelComment']['snippet']['likeCount'])
```

```python
video_name=req['items'][0]['snippet']['title']
channel_name=req['items'][0]['snippet']['channelTitle']
```

```python
#Making dictionary with all the data

yt_dict={
    'Video Name': video_name,
    'Channel Name':channel_name,
    'Comment Id':comment_id,
    'Name':comment_name,
    'Comment':comment,
    'Date':comment_date,
    'Likes':comment_likes
}
```

```python
#Making a dataframe from dictionary

import pandas as pd

df=pd.DataFrame(yt_dict)
```

```python
#Exporting the dataframe as csv

filename="{} {}.csv".format(channel_name,videoid)
df.to_csv(filename,index=False)
```
